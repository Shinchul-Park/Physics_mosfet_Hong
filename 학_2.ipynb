### 라플라스, 1차원
파동방정식: 고유값문제 H phi(x) = -k^2 * Delta_x^2 phi(x)로 Ax=lambda x로 고유값 문제가 된다.\
라플라스방정식: delta^2 phi(x) = 0 으로 Ax=b형태로 되어 선형방정식을 푸는 문제가 된다. \
만약 초기치가 x(0)=0, x(a)=1로 가정하면 아래와 같다.

```python
import numpy as np, matplotlib.pyplot as plt
from scipy.linalg import eigh

a=5 #length
N=5
Ndx=a/N
x=np.arange(N+1)*a/N
A=np.zeros((N+1,N+1))
b=np.zeros((N+1,1))
for ii in range(1,N):
    A[0,0]=1
    b[0]=0
    A[ii,ii-1]=1
    A[ii,ii  ]=-2 
    A[ii,ii+1]=1
    A[N,N]=1
    b[N]=1
phi=np.linalg.solve(A,b) #A행렬, b값을 주면, x(phi)를 풀어준다.
plt.plot(x,phi,'bo-')
plt.xlabel('position (nm)')
plt.ylabel('phi (1)')
plt.show()
```

### poisson equation
라플라스는 한번에 구할 수 있지만, poisson은 내부에 유전율이 존재하기 때문에 다르다.\
2계 미분항을 특정한 control volume에 대해서 적분해 주면, 2계 미분항이 1계 미분으로 바뀌게 된다.\
control volume은 i-0.5에서 i+0.5까지 적분해 준다.(interface)\
이렇게 하면, 이산화된 node에서 유전율이 계산되는 것이 아니라, + - 0.5에서 계산되기 때문에 문제가 없게 된다.\
알짜전하가 없는 경우

```python
import numpy as np
import matplotlib.pyplot as plt

a, e1, e2= 5, 11.7, 3.9 # 왜 N이 홀수면 이상해 지는가?
N=4
dx=a/N
x=np.arange(N+1)*dx # 이때 N=0~10번째 자리까지
A=np.zeros((N+1,N+1))
b=np.zeros((N+1,1))
for ii in range(1,N): # 이때 N=1~9번째 자리까지이다. 따라서 초기값 [0,0],[N,N]을 넣어주어야 한다.
    A[0,0]=1 
    if dx*ii<a/2:
        A[ii,ii-1], A[ii,ii  ], A[ii,ii+1]= e1, -2*e1,  e1
    elif dx*ii==a/2:
        A[ii,ii-1], A[ii,ii  ], A[ii,ii+1]= e1, -e2-e1, e2
    elif dx*ii>a/2:
        A[ii,ii-1], A[ii,ii  ], A[ii,ii+1]= e2, -2*e2,  e2
    A[N,N]=1
    b[N]=1 #potenital=1
    #print(A)

#print(b)
#print(A)
#print(b)
phi=np.linalg.solve(A,b)
#print(phi)
plt.plot(x,phi,'bo-')
plt.xlabel('position (nm)')
plt.ylabel('phi (1)')
plt.grid()
plt.show()
```

### poisson 방정식, 알짜전하가 있을 때
electron, hole이 없고, 단지 Ndop\
왜 N, width를 변경하면 풀리지 않을까?\
전체길이에 대해서 배수의 N이 되어야 풀린다.\
계속해서 interface에서의 전위가 변하고 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

q=1.6e-19
width=5e-9
tox=0.5e-9
Na=1e24
N=12 #(8+1*2)*4홀수가 되면 이상한 값이 된다. width와 tox를 달리하면 이상하다.
dx=(width+2*tox)/N
interface1=int(tox/dx) #0,1,2,3,---11 중 두번째 "1" '/','//'차이에도 계속변함.
interface2=int((tox+width)/dx) #0,1,2,---11,12 중 뒤에서 두번째 "11"
e1, e2, epsilon0 = 11.7, 3.9, 8.854e-12

x=np.arange(N+1)*dx 
A=np.zeros((N+1,N+1))
b=np.zeros((N+1,1))

for ii in range(1,N): 
    A[0,0]=1
    A[N,N]=1
    if interface1 <ii and ii<interface2:
        A[ii,ii-1], A[ii,ii], A[ii,ii+1]= e1, -2*e1,  e1
        b[ii]=dx*dx*q*Na/epsilon0
    elif ii<interface1 or ii>interface2:
        A[ii,ii-1], A[ii,ii], A[ii,ii+1]= e2, -2*e2,  e2
        b[ii]=0
    elif ii==interface1:
        A[ii,ii-1], A[ii,ii], A[ii,ii+1]= e2, -e2-e1,  e1
        b[ii]=dx*dx*q*Na/(2*epsilon0)
    elif ii==interface2:
        A[ii,ii-1], A[ii,ii], A[ii,ii+1]= e1, -e2-e1,  e2
        b[ii]=dx*dx*q*Na/(2*epsilon0)


b[0]=0 # b를 for 문 밖으로 보내야 한다.
b[N]=0

phi=np.linalg.solve(A,b)
#print(phi)

phi_interface=-3*tox*q*Na*width/(2*e1*epsilon0)
print(phi_interface)

plt.plot(x,phi,'bo-')
plt.xlabel('position (nm)')
plt.ylabel('phi (1)')
plt.grid()
plt.show()
```

effective density-of-states(DOS)를 먼저 알아야 한다.\

```python
import numpy as np
import matplotlib.pyplot as plt

# 기본 변수 설정
width = 5e-9  # 폭 (m)
tox = 0.5e-9  # 산화막 두께 (m)
Na = 1e24  # 도핑 농도 (/m³)
N = 24  # 계산 그리드 포인트 수
dx = (width + 2 * tox) / N  # 그리드 간격
interface1, interface2 = int(tox / dx), int((tox + width) / dx)  # 인터페이스 위치
q, epsilon0, e1, e2 = 1.6e-19, 8.854e-12, 11.7, 3.9  # 상수

# x 위치 배열
x = np.arange(N + 1) * dx

# 행렬 A와 벡터 b 초기화
A = np.zeros((N + 1, N + 1))
b = np.zeros(N + 1)

# 행렬 및 벡터 구성
for ii in range(1, N):
    if interface1 < ii < interface2:  # 반도체 영역
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e1, -2 * e1, e1
        b[ii] = dx * dx * q * Na / epsilon0
    elif ii < interface1 or ii > interface2:  # 산화막 영역
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e2, -2 * e2, e2
        b[ii] = 0
    elif ii == interface1:  # 첫 번째 인터페이스
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e2, -e2 - e1, e1
        b[ii] = dx * dx * q * Na / (2 * epsilon0)
    elif ii == interface2:  # 두 번째 인터페이스
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e1, -e1 - e2, e2
        b[ii] = dx * dx * q * Na / (2 * epsilon0)

# 경계 조건 설정
A[0, 0] = 1
A[N, N] = 1
b[0] = 0.33374  # 왼쪽 경계 전위
b[N] = 0.33374  # 오른쪽 경계 전위

# 전위 계산
phi = np.linalg.solve(A, b)

# 전자 밀도 및 정공 밀도 계산
nint = 1e10  # 내재 캐리어 농도 (/cm³)
kB = 1.38065e-23  # 볼츠만 상수 (J/K)
T = 300  # 온도 (K)

ec = np.zeros(N + 1)
hc = np.zeros(N + 1)
for ii in range(N + 1):
    if ii<interface1 or ii>interface2: 
        ec[ii], hc[ii] = 0, 0
    else:
        ec[ii] = nint * np.exp(q * phi[ii] / (kB * T))
        hc[ii] = nint * np.exp(-q * phi[ii] / (kB * T))

# 결과 시각화
plt.plot(x * 1e9, ec, 'bo-', label='Electron density')
plt.xlabel('Position (nm)')
plt.ylabel('Electron density (/cm³)')
plt.title('Electron Density Profile')
plt.grid()
plt.legend()
plt.show()
```

### 2.10 Newton 방법
비선형방정식 수치해석 - Newton Raphson\
접선의 기울기 x0에서의 기울기 m=f(x0)/(x0-x1), x1=x0-f(x0)/f'(x0) -> x1에서 기울기, x2에서의 기울기를 가지고 계속 수행해서 해를 찾는다. \
첫번째 농도를 계산하기 위해 increment, iteration, 인정(convergence,수렴), 두번째 농도 포인트 increment2, iteration, convergence...\
행렬에 적용하면, Jacobian이 된다.\

평형상태: ec,hc potential 함수가 가능\
대신 poisson 방정식의 알짜 전하량(Net charge)이 비선형함수가 된다. 이것을 풀어야 한다.

```python
#f(x)=x^2-1, f'(x)=2x,f(x)=0를 구하자...

import numpy as np
x0=2 # intical value
print(x0)
for i in range(0,10):
    fx0=pow(x0,2)-1
    dfdx=2*x0
    deltax=-fx0/dfdx
    x0=x0+deltax
    print(x0)
```

```python
#평형상태에서의 phi 만 알면 전자농도, 홀농도는 된다.
#도핑농도가 주어졌을 때 charge neutrality를 풀어야 한다.
#하나의 변수에 대한 하나의 비선형 방정식을 푼다.

import numpy as np
import matplotlib.pyplot as plt

# 상수 정의
q, epsilon0 = 1.602e-19, 8.854e-12
e1, e2 = 11.7, 3.9
nint, kB, T = 1e16, 1.38065e-23, 300
thermal = kB * T / q  # 열전위 계산

# 변수 초기화
phi_numeric = np.zeros(8)  # 수치적으로 구한 phi
phi_analytic = np.zeros(8)  # 분석적으로 구한 phi

# Newton-Raphson 반복 시작
for iDop in range(8):
    Ndop = 10 ** (15 + iDop)  # 도핑 농도
    phi = 0.3  # 초기 guess
    for Newton in range(39):  # 최대 39번 반복
        #for Newton in range(39) 구문은 반복 횟수만 제한하기 위한 용도로 사용되고 있고, 실제로 Newton 변수는 계산에 사용되지 않습니다.
        # 잔차 및 야코비 행렬
        res = q * (Ndop + nint * np.exp(-phi / thermal) - nint * np.exp(phi / thermal)) #fx0, residue vector가 된다.
        Jaco = q * (-nint * np.exp(-phi / thermal) - nint * np.exp(phi / thermal)) / thermal #dfdx, 부호를 주의해야 한다.
        update_phi = -res / Jaco  # delta_phi=-fx0/dfdx Newton-Raphson 업데이트 
        phi += update_phi  # phi=phi+delta_phi phi 갱신 
                
        if abs(update_phi) < 1e-15:  # 수렴 조건, delta_phi
            break

    phi_numeric[iDop] = phi  # 수치적으로 구한 값 저장
    phi_analytic[iDop] = thermal * np.arcsinh(Ndop / (2 * nint))  # 분석적 해

# 결과 출력
plt.plot(range(15, 23), phi_numeric, 'bo-', label='Numerical Solution')
plt.plot(range(15, 23), phi_analytic, 'r--', label='Analytical Solution')
plt.xlabel('log10(Ndop) (/cm³)')
plt.ylabel('Potential (V)')
plt.title('Comparison of Numerical and Analytical Solutions')
plt.legend()
plt.grid()
plt.show()
```

### Newton-Raphson, Jacobian

```python
import numpy as np
import matplotlib.pyplot as plt


# Initial guess for phi
phi = np.array([1, 2, 3])

# Maximum number of iterations
max_iterations = 10

# Perform Newton-Raphson iterations
for newton in range(max_iterations):
    # Construct the Jacobian matrix
    jaco = np.array([[-2 - np.exp(phi[0]), 1, 0], [1, -2 - np.exp(phi[1]), 0], [0, 1, -2 - np.exp(phi[2])]])

    # Compute the residual vector
    res = np.array([phi[1] - 2*phi[0] - np.exp(phi[0]), phi[2] - 2*phi[1] + phi[0] - np.exp(phi[1]), 4 - 2*phi[2] + phi[1] - np.exp(phi[2])])

    # Solve for the update using the Jacobian and residual
    update = np.linalg.solve(jaco, -res)

    # Update phi
    phi = phi+update

print(phi)

# Plot the final values of phi
plt.plot(phi,'bo-')
plt.xlabel('Iteration')
plt.ylabel('Phi values')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 기본 변수 설정
width = 5e-9  # 폭 (m)
tox = 0.5e-9  # 산화막 두께 (m)
Na = 1e24  # 도핑 농도 (/m³)
N = 60  # 계산 그리드 포인트 수
dx = (width + 2 * tox) / N  # 그리드 간격
interface1, interface2 = int(tox / dx), int((tox + width) / dx)  # 인터페이스 위치
q, e0, e_si, e_ox = 1.6e-19, 8.854e-12, 11.7, 3.9  # 상수
nint = 1e16  # 내재 캐리어 농도 (/m³)
kB = 1.38065e-23  # 볼츠만 상수 (J/K)
T = 300  # 온도 (K)

# x 위치 배열
x = np.arange(N + 1) * dx
coeff = dx * dx * q / e0
thermal = kB * T / q

Elec_save = np.zeros((N + 1, 101))

# 행렬 A와 벡터 b 초기화
A = np.zeros((N + 1, N + 1))
b = np.zeros(N + 1)

# 행렬 및 벡터 구성
for ii in range(1, N):
    if interface1 < ii < interface2:  # 반도체 영역
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e_si, -2 * e_si, e_si
        b[ii] = coeff * Na
    elif ii < interface1 or ii > interface2:  # 산화막 영역
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e_ox, -2 * e_ox, e_ox
        b[ii] = 0
    elif ii == interface1:  # 첫 번째 인터페이스
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e_ox, -e_ox - e_si, e_si
        b[ii] = coeff * Na / 2
    elif ii == interface2:  # 두 번째 인터페이스
        A[ii, ii - 1], A[ii, ii], A[ii, ii + 1] = e_si, -e_si - e_ox, e_ox
        b[ii] = coeff * Na / 2

# 경계 조건 설정
A[0, 0] = 1
A[N, N] = 1
b[0] = 0.33374  # 왼쪽 경계 전위
b[N] = 0.33374  # 오른쪽 경계 전위

# 초기 전위 계산
phi = np.linalg.solve(A, b)

# 게이트 전압 설정

Vg_array = np.linspace(0, 1, 11)  # From 0V to 1V in steps of 0.1V
Electron_array = []

for index_Vg, Vg in enumerate(Vg_array): 
    
    # 뉴턴 반복
    for Newton in range(20):
        jaco = np.zeros((N + 1, N + 1))
        res = np.zeros(N + 1)
        for ii in range(N+1):
            if ii == 0 or ii == N:  # 경계 조건
                res[ii] = phi[ii] - 0.33374 - Vg
                jaco[ii, ii] = 1
            elif interface1 < ii < interface2:  # 반도체 영역
                res[ii] = e_si * (phi[ii - 1] - 2 * phi[ii] + phi[ii + 1]) + coeff * (-Na - nint * np.exp(phi[ii] / thermal))
                jaco[ii, ii - 1] = e_si
                jaco[ii, ii] = -2 * e_si - coeff * nint * np.exp(phi[ii] / thermal) / thermal
                jaco[ii, ii + 1] = e_si
            elif ii < interface1 or ii > interface2:  # 산화막 영역
                res[ii] = e_ox * (phi[ii - 1] - 2 * phi[ii] + phi[ii + 1])
                jaco[ii, ii - 1] = e_ox
                jaco[ii, ii] = -2 * e_ox
                jaco[ii, ii + 1] = e_ox
            elif ii == interface1:  # 첫 번째 인터페이스
                res[ii] = e_ox * (phi[ii - 1] - phi[ii]) + e_si * (phi[ii + 1] - phi[ii]) + coeff * (-Na - nint * np.exp(phi[ii] / thermal)) / 2
                jaco[ii, ii - 1] = e_ox
                jaco[ii, ii] = -e_ox - e_si - coeff * nint * np.exp(phi[ii] / thermal) / (2 * thermal)
                jaco[ii, ii + 1] = e_si
            elif ii == interface2:  # 두 번째 인터페이스
                res[ii] = e_si * (phi[ii - 1] - phi[ii]) + e_ox * (phi[ii + 1] - phi[ii]) + coeff * (-Na - nint * np.exp(phi[ii] / thermal)) / 2
                jaco[ii, ii - 1] = e_si
                jaco[ii, ii] = -e_si - e_ox - coeff * nint * np.exp(phi[ii] / thermal) / (2 * thermal)
                jaco[ii, ii + 1] = e_ox

        # Singular matrix 방지
        if np.any(np.abs(np.diag(jaco)) < 1e-10):
            print(f"Iteration {Newton}: Near-zero diagonal element in Jacobian matrix.")
            break

        # Newton-Raphson 갱신
        update = np.linalg.solve(jaco, -res)
        phi += update

        # 수렴 검사
        if np.linalg.norm(update) < 1e-10:
            print(f"Converged at iteration {Newton}")
            break

    
    # Elec_save 계산
    Elec_save[:, index_Vg] = nint * np.exp(phi / thermal) *1e-6

    # Electron 밀도 계산
    Electron = (
        Elec_save[interface1, index_Vg] * dx * 100 / 2 +
        np.sum(Elec_save[interface1 + 1:interface2, index_Vg] * dx * 100) +
        Elec_save[interface2, index_Vg] * dx * 100 / 2
    )
    Electron_array.append(Electron)

plt.figure(figsize=(8,4))
plt.subplot(121)
plt.plot(Vg_array, Electron_array, 'bo-', label='Electron Density')
plt.xlabel('Gate Voltage (V)'), plt.ylabel('Electron Density (cm⁻³)')
plt.ylim([1e9,6e13])
plt.grid(), plt.legend()

plt.subplot(122)
plt.semilogy(Vg_array, Electron_array, 'bo-', label='Electron Density')
plt.xlabel('Gate Voltage (V)'), plt.ylabel('Electron Density (cm⁻³)')
plt.ylim([1e9,1e14])
plt.grid(), plt.legend()

plt.tight_layout(), plt.show()    
```
